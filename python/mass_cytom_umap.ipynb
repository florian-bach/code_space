```python
len.(flow_data.columns)
cytobank = flow_data.iloc[:,4:56]
cytobank.head(20)
```

```python
import umap
reducer = umap.UMAP()

embedding = reducer.fit_transform(cytobank)
embedding.shape

    
```

```python
list(cytobank.columns.values)
```

```python
cytobank['umap1'] = embedding[:, 0]
cytobank['umap2'] = embedding[:,1]
```

```python
channels=list(cytobank.columns.values)

for n in channels:
    cmap=plt.get_cmap('nipy_spectral')
    fig=plt.figure(figsize=(15,10))
    ax=plt.subplot()
    ax.scatter(cytobank['umap1'], cytobank['umap2'],c=np.arcsinh(cytobank[n]),
               cmap=cmap, s=1)
    plt.title('{}'.format(n))
    #plt.savefig('{}.pdf'.format(n))
    plt.show()
```
